# Removing HTML Tags

* The re library in Python is the built-in module for working with regular expressions (regex). Regular expressions are patterns that provide a powerful way to search, match, and manipulate text.

* In this class remove_html_tags is applied on the text and the in the next step the pattern stores the words which has matching expression such as '<.*?>' in the text and then after processing it removes the expressions and return the text with an empty string 

```python
import re  #Regular Expressions
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

* Now by applying the below code the html tags get removed provide an empty string by replacing expression.

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
df
```

# Removal of url 

* Now a class is applied on text named remove_url it also performs same as html tags removal and returns text in plain form.

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www.\.\S+')
    return pattern.sub(r'', text)
```

```python
df['review'] = df['review'].apply(remove_url)
```

```python
df
```

# Removal of Punctuations

* in the below code 
  string: Provides constants, classes, and functions for string operations; 
  time: Provides functions to work with time-related operations (used for delays, timestamps, etc.).
* string.punctuation is a constant that contains a string of all printable punctuation characters.


```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))

df['review'] = df['review'].apply(remove_punc)
```

```python
df
```

# Removal of Stop Words

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append(' ')

        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return" ".join(x)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```
