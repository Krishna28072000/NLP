# Applying Stemming

Stemming is the process of reduction of words in to its base or root word known as stem, the goal of stemming is to remove prefixes , suffixes and other to get the word root form.

* Stemming is also categorized in to 3 types

1. Poter Stemmer : It uses sequences of steps and applies specific rules which remove common suffixes from the words and reducing the word into stem.

2. Lancaster Stemmer : It uses sequence of steps and applies same logic as porete stemmer but it is more aggressive so it reduce words into its very basic form.
Ex : Running = run: easily = easili

3. Snowball Stemmer : It  applies a set of rules to remove suffixes but is more accurate and provides better handling of edge cases. It is more extensive and supports several languages.
Ex : Running = run

* word_tokenize() is a function from the NLTK library that tokenizes a sentence into words. This means it breaks a sentence into individual words and punctuation marks.
It’s a method that first splits the sentence into tokens (words and punctuation) and returns them as a list.

* for w1 in word_tokens::

This is a for loop that iterates over each word (or token) in the word_tokens list, which was created in Code 1.
w1 represents a single word or token in each iteration.

* porter_stemmer.stem(w1):

porter_stemmer is an instance of the Porter Stemmer algorithm from the NLTK library. The stem() method is applied to each word (w1).
Stemming is the process of reducing a word to its base or root form by removing suffixes. The Porter Stemmer applies a series of rules to strip suffixes and reduce words to their stem form.

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
word_tokens = word_tokenize(sent_tokenize_list[0])
```

```python
st_list=[]
for w1 in word_tokens:
    st_list.append(porter_stemmer.stem(w1))
```

```python
[porter_stemmer.stem(test) for test in word_tokens]
```
